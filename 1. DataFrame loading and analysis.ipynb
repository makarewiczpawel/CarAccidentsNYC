# 2. Data cleaning and preparation

```python
# Zamiana typu danych w kolunnie 3 [ZIP CODE] na 'str'
vehicle_colision_cleaned = pd.read_csv('DataSets/nypd-motor-vehicle-collisions.csv',
                                      dtype={'ZIP CODE': str}
                                      )
vehicle_colision_cleaned.isnull().sum()
```

### Usuwanie duplikatów

```python
# Wyświetlenie ramki danych z duplikatami po kolumnie "COLLISION_ID"
```

```python
# Usunięcie duplikatów na podstawie kolumny 'COLLISION_ID'
display(f"Before drop_duplicates: {len(vehicle_colision_cleaned)}")
display(f"After drop_duplicates: {len(vehicle_colision_cleaned.drop_duplicates(subset=['COLLISION_ID']))}")
vehicle_colision_cleaned = vehicle_colision_cleaned.drop_duplicates(subset=['COLLISION_ID'])
```

### Usuwanie pustych wierszy

```python
# Brak pustych wierszy w kolumnie "COLLISION_ID"
vehicle_colision_cleaned['COLLISION_ID'].isnull().sum()
```
